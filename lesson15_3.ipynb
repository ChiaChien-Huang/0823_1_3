```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

student1= {'':78,'':80,'':92,'':75,'':85}
student2= {'':90,'':75,'':95,'':65,'':75}
#windows
#font=FontProperties(fname=r'c:windows\fonts\simsun.ttc',size=8)
#mac
#font= FontProperties(fname='/Library/Fonts/Arial Unicode.ttc',size=8)

#codespaceLinux
font= FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)

subjects= list(student1.keys())
value1= list(student1.values())
value2= list(student2.values())

figure= plt.figure()
axes=figure.add_subplot(1,1,1)
axes.plot(subjects,value1,'m-.o')
axes.plot(subjects,value2,'c-.o')
axes.set_xticklabels(subjects,fontproperties=font)
axes.set_title('12',fontproperties=font)
axes.grid(axis='y')
axes.legend(["1","2"],prop=font)
x_loc= axes.get_xticks()

for index,x in enumerate(x_loc):
    y= value1[index]
    axes.annotate(str(y),xy=(x,y))

for index,x in enumerate(x_loc):
    y= value2[index]
    axes.annotate(str(y),xy=(x,y))
plt.show()
```
