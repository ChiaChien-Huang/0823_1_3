```python
import matplotlib.pyplot as plt
values= [20,30,45,10]
label= ['Nokia','Samsung','Motorola','Apple']
color= ['yellow','green','red','blue']

figure=plt.figure()
axes=figure.add_subplot(1,1,1)
axes.pie(values,labels=label,colors=color,explode=[0.3,0,0,0],shadow=True,startangle=180,autopct='%.1f%%')
plt.show()
```
