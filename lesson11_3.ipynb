```python
class Student:
    def __init__(self,name:str,ch:int,eng:int,math:int,na:int,so:int):
        #attribute
        self.姓名=name
        self.國文=ch
        self.英文=eng
        self.數學=math
        self.自然=na
        self.社會=so

    def __repr__(self) -> str:
        return f'{self.姓名}的實體'
    
    #總分實體方法
    def sum(self) ->int:
        return self.國文+self.英文+self.數學+self.自然+self.社會
    
    #平均實體方法
    def avg(self) ->float:
        return round(self.sum()/5, ndigits=2)
        
```

```python
import csv
with open("學生成績表1.csv",encoding='utf-8',newline='') as file:
    dict_reader=csv.DictReader(file)
    students=[]
    for student in dict_reader:
        chinese=int(student['國文'])
        english=int(student['英文'])
        math=int(student['數學'])
        nature=int(student['自然'])
        social=int(student['社會'])
        student=Student(student['姓名'],
                        ch=chinese,
                        eng=english,
                        math=math,
                        na=nature,
                        so=social)
        students.append(student)
        
        
        
students
```

```python
for stu in students:
    print(stu.姓名)
    print(f'總分:{stu.sum()}')
    print(f'平均:{stu.avg()}')
    print("======================")
```
