```python
import matplotlib.pyplot as plt

stu1={'chinese':80, 'english':90, 'math':85, 'history':75,'geography':65}
stu2={'chinese':60, 'english':85, 'math':62, 'history':73,'geography':95}
subjects=list(stu1.keys())
s1_value=list(stu1.values())
s2_value=list(stu2.values())


flg=plt.figure(figsize=(8,5))
axes1= flg.add_subplot(2,1,1)
axes1.plot(subjects,s1_value,'b-.o')
axes1.plot(subjects,s2_value,'r-.o')
axes1.grid(axis='y')

axes2= flg.add_subplot(2,1,2)
axes2.plot(subjects,s1_value,'b-.o')
axes2.plot(subjects,s2_value,'r-.o')
axes2.grid(axis='y')
xmin,xmax,ymin,ymax =axes2.axis()
axes2.axis(xmin=1.5,xmax=2.5,ymin=ymin,ymax=ymax+10)
#print(axes.get_xticks())
#print(axes.get_xticklabels())
plt.show()
```
