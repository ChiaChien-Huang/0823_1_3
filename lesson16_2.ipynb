```python
import numpy as np
import pandas as pd

scores=np.random.randint(50,high=101,size=(50,5))
dataf=pd.DataFrame(scores,
             columns=["國文","英文","數學","地理","歷史"],
             index=range(1,51))
dataf.index.name="學號"
dataf.columns.name="科目"
dataf
```

```python
sumvalues=dataf.sum(axis=1)
avgvalues=dataf.mean(axis=1)
dataf["總分"] = sumvalues
dataf["平均"] = avgvalues
dataf
```
