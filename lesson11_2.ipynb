```python
import csv
with open("學生成績表1.csv",encoding='utf-8',newline='') as file:
    dict_reader=csv.DictReader(file)
    students=list(dict_reader)
    for student in students:
        student['國文']=int(student['國文'])
        student['英文']=int(student['英文'])
        student['數學']=int(student['數學'])
        student['自然']=int(student['自然'])
        student['社會']=int(student['社會'])
students
        
```

```python
for stu in students:
    print(stu['姓名'])
    sum= stu['國文']+stu['英文']+stu['數學']+stu['自然']+stu['社會']
    print(f"總分{sum}")
    print(f"平均{sum/5:.2f}")
    print("====================")
```
