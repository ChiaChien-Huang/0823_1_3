```python
import csv

with open("學生成績表1.csv",encoding="utf-8", newline="") as file:
    csvReader=csv.reader(file)
    next(csvReader)
    #for row in csvReader:
    #    print(row)
    students=list(csvReader)
students
```

```python
print("國文>=80的學生有:")
amount=0
for student in students:
    amount+=1
    print(f"學生:{student[0]},國文{student[1]}")
print(f"共有{amount}學生")
```
